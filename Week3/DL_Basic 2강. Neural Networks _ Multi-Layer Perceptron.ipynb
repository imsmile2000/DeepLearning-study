## Neural Networks & Multi-Layer Perceptron

**Neural Network**
- 애매하게 뇌를 모방한 computing 시스템

**Linear Neural Networks**
- 목적: 입력이 1차원이고 출력이 2차원일때 이 두개를 연결하는 모델을 찾는 것
- $ŷ=wx+b$에서 w,b 찾기
- 새로운 w와 b는 stepsize*(w와b의 편미분값)을 w,b에서 빼준 것
image.png


$y = W^T x + b ⟺ x - (W,b) → y$
- W는 행렬 b는 벡터

### Activation functions
- 비선형(Activation) 함수가 들어가야지만 네트워크를 쌓았을 때 의미가 있음
- ReLU, Sigmoid, Hyperbolic Tangent

## Multi-Layer Perceptron
- 다층 구조의 퍼셉트론
  -$y=W_3^Th_2=W_3^Tρ(W_2^Th_1)$
  $=W_3^Tρ(W_2^Tρ(W_1^Tx))$
- loss functions
  - Regression Task (MSE): 회귀 문제 풀 때
  - Classification Task (CE): 분류 문제 풀 때
  - Probabilistic Task(MLE): 단순히 숫자 값으로 결과가 나오는게 아니라 불확실한 문제 풀 때

## Further Questions

1. Regression Task와 Classification Task의 loss function이 다른 이유는 무엇인가요?
2. Regression Task, Classification Task, Probabilistic Task의 Loss 함수(or 클래스)는 Pytorch에서 어떻게 구현이 되어있을까요? 
 - Regression Task
  ```
  loss = nn.MSELoss()
  input = torch.randn(3, 5, requires_grad=True)
  target = torch.randn(3, 5)
  output = loss(input, target)
  output.backward()
  ```
   - Classification Task
  ```
  # Example of target with class indices
loss = nn.CrossEntropyLoss()
input = torch.randn(3, 5, requires_grad=True)
target = torch.empty(3, dtype=torch.long).random_(5)
output = loss(input, target)
output.backward()
# Example of target with class probabilities
input = torch.randn(3, 5, requires_grad=True)
target = torch.randn(3, 5).softmax(dim=1)
output = loss(input, target)
output.backward()
  ```
