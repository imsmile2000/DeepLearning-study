## 1. 모수
- **통계적 모델링은 적절한 가정 위에서 확률분포를 추정**하는 것이 목표
- 유한한 개수의 데이터로는 분포를 정확하게 알아낼 수 없다. **근사적으로 확률분포를 추정**하는 것일 뿐!

#### 방법론 - 확률분포 가정 여부에 따라 달라짐
- **모수적 방법론**: 데이터가 특정 분포를 따른다고 가정한 후 그 분포를 결정하는 모수를 추정
- **비모수 방법론**: 특정 확률분포를 가정하지 않고 데이터에 따라 모델 구조와 모수의 개수가 바뀜


## 2. 확률분포 가정하기
1. 데이터가 0또는 1만 가지는 경우 &rarr; `베르누이분포`
2. 데이터가 n개의 이산적인 값을 가지는 경우 &rarr; `카테고리분포`
3. 데이터가 [0,1] 사이에서 값을 가지는 경우 &rarr; `베타분포`
4. 데이터가 0 이상의 값을 가지는 경우 &rarr; `감마분포,로그정규분포 등`
5. 데이터가 $R$ 전체에서 값을 가지는 경우 &rarr; `정규분포,라플라스분포 등`

### 3-1. 최대가능도 추정법 (MLE)
- **이론적으로 가장 가능성이 높은** 모수 추정법

    > argmax (가능도 함수): 확률이 아님

- 데이터 집합 X가 독립적으로 추출되었을 경우, **로그가능도**를 최적화
    - 로그 가능도: 가능도 함수의 로그
    
    <br></br>
    - 로그 가능도를 사용하는 이유 `최적화 때문!`
        1. 데이터 숫자가 엄청 커지면 컴퓨터의 정확도로 가능도 계산 불가
        2. 데이터가 독립이면 가능도의 곱셈을 로그가능도의 덧셈으로 바꿀 수 있음 
        3. 연산량: 경사하강법 $O(n^2)$, 로그가능도 $O(n)$

## Further Question
1. 확률과 가능도의 차이는 무엇일까요? (개념적인 차이, 수식에서의 차이, 확률밀도함수에서의 차이)
> - 확률: 주어진 확률분포에서 특정 결과가 나올 가능성
> $P~r(a<=X<=b)=\int_a^b \mathrm f(x)\,\mathrm{d}x$
> - 가능도: 주어진 결과값이 특정 확률분포로부터 나왔을 확률
> $P~r(X=x|θ)=f(x)$

> 위에가 확률, 아래가 가능도
    

2. 확률 대신 가능도를 사용하였을 때의 이점은 어떤 것이 있을까요?
> 주어진 데이터를 가지고 모수를 추정할 수 있음
3. 다음의 code snippet은 어떤 확률분포를 나타내는 것일까요? 해당 확률분포에서 변수 theta가 의미할 수 있는 것은 무엇이 있을까요?
> $f(θ)=θ^3-(1-θ)^7$

```python
#n=10, k=3, p=θ인 이항분포
import numpy as np
import matplotlib.pyplot as plt
theta = np.arange(0, 1, 0.001)
p = theta ** 3 * (1 - theta) ** 7
plt.plot(theta, p)
plt.show()
```
