## Convolution 연산
- 다층신경망(MLP): 각 뉴런들이 선형모델과 활성함수로 모두 연결된 구조
- Convolution 연산은 이와 달리 **커널(kernel)을 입력벡터 상에서 움직여가면서** 선형모델과 합성함수가 적용되는 구조
- Convolution 연산은 **선형변환** (활성화 함수 제외)

- 수학적인 의미: 신호($g(x)$)를 커널($f(x)$)를 이용해 국소적으로 증폭/감소시켜 정보를 추출 또는 필터링 하는 것 $\rightarrow [f*g](x)$
- 커널은 정의역 내에서 움직여도 변하지 않고 주어진 신호에 국소적으로 적용

### 2차원 Convolution 연산
- $[f*g](i)=\sum_{p=1}^df(p)g(i+p)$
- ex)
    

   - 첫번째 칸: $0*0+1*1+2*3+3*4=19$
   - 두번째 칸: $0*1+1*2+3*4+3*5=25$

- 입력 크기: $(H,W)$ 
- 커널 크기: $(K_{H},K_{W})$
- 출력 크기: $(O_{H},O_{W})$ 라고 하면
- 다음 출력 크기
     $$ O_{H}=H-K_{H}+1 $$
     $$ O_{W}=W-K_{W}+1 $$
     
<br></br>
- 커널이 하나면 $(O_{H},O_{W},1)$이지만 여러개 쌓으면 $(O_{H},O_{W},O_{C})$

## Convolution 연산의 역전파
- 역전파도 convolution 연산
- ex)
    입력=5, 커널=3 이면 출력=5-3+1=3

#### 느낀 점
- convolution 연산은 전공 통신 수업을 들으면서 여러번 접해봤던 내용이여서 크게 어렵지는 않았다. 연산은 어렵지 않지만 이론적인 부분들은 처음 접해본 내용도 있어서 이론 부분에 조금 더 신경써야할 것 같다
