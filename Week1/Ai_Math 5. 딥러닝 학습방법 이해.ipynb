## 신경망
- 비선형모델
- 각 행벡터 데이터와 가중치행렬(W) 사이의 행렬곱과 절편의 합으로 표현됨
- 신경망은 **선형 모델과 활성함수를 합성한 함수**

### 소프트맥스 연산
- 모델의 출력을 확률로 해석할 수 있게 변환해주는 연산
- 분류 문제를 풀 떼 선형모델과 소프트맥스 함수 결합
- 추론을 할 때는 **one-hot 벡터**로 최대값을 가진 주소만 1로 출력 (소프트맥스 사용 안함)

### 활성함수
- **비선형함수**로서 딥러닝에서 매우 중요한 개념
- 활성함수를 쓰지 않으면 딥러닝은 선형모형과 차이 없음
- 시그모이드 / $tanh$함수(전통)   **ReLU (가장 많이 쓰임)**
- $tanh(x)={(e^{2x}-1)}/{(e^{2x}+1)}$
<br></br>
- $tanh'(x)=1-tanh^2(x)$

## 신경망 쌓는 과정
1. 활성함수 σ가 잠재벡터(z)의 각 노드에 개별적으로 적용하여 새로운 잠재벡터(H)를 만듦


2. 잠재벡터(H)에서 가중치 행렬 $W^{(2)}$와 $b^{(2)}$를 통해 다시 한번 선형변환하면 $(W^{(2)},W^{(1)})$를 패러미터로 가진 **2-layers 신경망**
    

3. 반복하면 **다층 퍼셉트론(MLP)**: 신경망이 여러층 합성된 함수
    

### 층을 여러개 쌓는 이유
- 이론적으로는 2-layers 신경망으로도 연속함수 근사 가능, 하지만 성능이 안좋다
- 층이 깊을수록 **목적함수를 근사하는데 필요한 노드의 숫자가 훨씬 빨리 줄어들어 더 효율적으로 학습 가능**

## 역전파 알고리즘 (backpropagation)
- 딥러닝은 역전파 알고리즘을 이용하여 각 층에 사용된 패러미터 $(W^{(i)},b^{(i)})$학습
- 각 층 패러미터의 그레디언트 벡터는 윗층부터 역순으로 계산
    * 합성함수 미분법인 **연쇄법칙 기반 자동 미분** 사용
    
                        -> 그래서 메모리를 순전파보다 많이 잡아먹음

### Further Question
Q. 분류 문제에서 softmax 함수가 사용되는 이유는?
> A. 모델의 결과를 확률로 표현하여 분류 결과와 비교할 수 있기 때문!

Q. softmax 함수의 결과값을 분류 모델의 학습에 어떤식으로 사용할 수 있을까요?
> A. softmax 함수의 결과인 **예측값**과 **실제값**의 오차를 최소화하는 방법으로 학습에 사용
