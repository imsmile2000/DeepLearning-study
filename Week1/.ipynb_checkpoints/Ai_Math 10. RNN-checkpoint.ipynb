## 시퀀스 데이터
- **시간 순서에 따라 나열된 데이터** ex)소리, 문자열, 주가, 시계열 등
- 독립동등분포 가정을 잘 위배함(순서를 바꾸거나 과거 정보가 손실되면 데이터의 확률분포도 바뀜)


## 시퀀스 데이터 다루는 방법
- **조건부 확률 이용**
$$P(X_{1},....X_{t})$$
$$=P(X_{t}|X_{1},....X_{t-1})P(X_{1},....X_{t-1})$$
$$=P(X_{t}|X_{1},....X_{t-1})P(X_{t-1}|X_{1},....X_{t-2})P(X_{1},....X_{t-2})$$
$$...=\prod_{s=1}^tP(X_{s}|X_{s-1},....X_{1})$$
- 모든 과거 정보들이 필요한 것은 아님

- 길이가 가변적인 데이터를 다룰 수 있는 모델이 필요
    $$X_{t}~P(X_{t}|X_{t-1},....X_{1})$$
    $$X_{t+1}~P(X_{t+1}|X_{t},....X_{1})$$
    <br></br>
    - 고정된 길이만큼의 시퀀스만 사용하는 경우 **자기회귀모델**이라고 부름
    - 바로 이전 정보를 제외한 나머지 정보들을 사용하는 경우 **잠재AR 모델**이라고 부름
    - 잠재변수 $H_{t}$를 신경망을 통해 반복해서 사용하여 시퀀스 데이터의 패턴을 학습하는 모델을 **RNN**이라고 부름

## RNN
- 가장 기본적인 RNN 모형은 MLP와 유사
- RNN은 **이전 순서의 잠재변수($H_{t-1}$)와 현재의 입력($X_{t})$**을 활용
$$O_{t}=H_{t}W^{(2)}+b^{(2)}$$
$$H_{t}=σ(X_{t}W_{X}^{(1)}+H_{t-1}W_{H}^{(1)}+b^{(1)})$$
<br></br>
- $W$는 시퀀스(t)와 상관없이 불변

## BPTT (RNN의 역전파)
- 잠재변수의 연결그래프에 따라 순차적으로 계산
- BPTT를 통해 RNN의 각 가중치행렬의 미분을 계산해보자
    
<br></br>
- 시퀀스의 길이가 길어지는 경우 $\rightarrow$ BPTT 계산이 불안정해짐 $\rightarrow$ 그레디언트가 0이되는 기울기 소실이 일어남 $\rightarrow$ 과거 정보를 유실 $\rightarrow$ **길이를 끊어줘야함** (truncated BPTT)

- truncated BPTT: 그레디언트를 나눠서 전달하는 방식, 기울기 소실을 해결

- Vanilla RNN은 길이가 긴 시퀀스를 처리하는데 문제가 있음 $\rightarrow$ LSTM이나 GRU 사용

#### 새롭게 알게된 사실
- RNN과 시퀀스 데이터에 대해서는 접해본적이 있는데 BPTT에 대해서는 처음 알게 되었다. 그레디언트 소실을 방지하기 위해 truncated BPTT를 사용한다는 점이 흥미로웠던 것 같다.

#### 일주일 회고
- 일주일동안 정말 많은 정보들을 접하게 되어서 한학기 강의를 압축해서 듣는 기분이었다.. 아직 100% 흡수된것 같지는 않아서 주말에 복습을 해야할 것 같다.
