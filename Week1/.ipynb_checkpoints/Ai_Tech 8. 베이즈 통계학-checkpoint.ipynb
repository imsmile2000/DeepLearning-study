## 조건부 확률
- $P(A \cap B)$=$P(B)P(A|B)$
- $P(B|A)$= $P(B)P(A|B) \over P(A)$

## 베이즈 정리
- 사후확률 $P(θ|A)= P(θ)$ $P(D|θ) \over P(D)$

### 예제
- COVID-99 의 발병률이 10% 로 알려져있다. COVID-99 에 실제로 걸렸을
때 검진될 확률은 99%, 실제로 걸리지 않았을 때 오검진될 확률이 10% 라고
할 때, **어떤 사람이 질병에 걸렸다고 검진결과가 나왔을 때 정말로 COVID-99 
에 감염되었을 확률**은?
    - 사전 확률 $P(θ)=0.1$
    - 가능도 $P(D|θ)=0.99$
    
    - evidence $P(D)= \sum P(D|θ)P(θ)$ =0.99\*0.1+0.1*0.9=0.189
    
    - 정답: $P(θ|D)=0.1$*$0.99 \over 0.189$=0.524

- <font color=blue>**True Positive**</font>: 양성이 나왔을 때 실제 양성인 경우
- <font color=blue>**False Positive**</font>: 음성이 나왔을 때 실제 음성일 경우

### 베이즈 정리 - 정보의 갱신
- $P(θ|D)$를 다시 사전 확률로 사용하여 갱신된 사후확률 계산
- 앞서 COVID-99 판정을 받은 사람이 두 번째 검진을 받았을 때도 양성이 나왔을
때 진짜 COVID-99 에  걸렸을 확률은?
    - 사전확률 $P(θ|D)=0.524$
    - 가능도 $P(D|θ)=0.99$
    - evidence $P(D')=0.99*0.524+0.1*0.476=0.566$
    - 정답: $P(θ|D')=0.524$*$0.99 \over 0.566$=0.917

## 조건부 확률로 인과관계를 추론?
- 조건부 확률은 통계적으로 유용한 결과를 제공하지만, **인과 관계 추론에서 함부로 사용하면 안됨!**
- 인과관계는 **데이터 분포의 변화에 강건함**, 하지만 정확도는 떨어짐
- 인과관계를 알아내려면 **중첩요인의 효과를 제거하고** 원인에 해당하는 변수만의 인과관계를 계산해야함
